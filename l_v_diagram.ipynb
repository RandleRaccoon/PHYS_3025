```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np
```

```python
R0 = 8 #kpc
v0 = 220 #km/s
```

```python
y_arr = []
x_arr = []
for R in range(1,20):
  for theta in range(0,180,10):

    l = np.arctan2(np.sin(theta*(np.pi/180)),(R0/R - np.cos(theta*(np.pi/180))))
    v = (v0/R - v0/R0)*R0*np.sin(l)

    #print(l)
    x_arr.append(l/((np.pi/180)))
    y_arr.append(v)
    #print(l,v)
print(len(x_arr))
print(len(y_arr))
```

```python
fig,ax=plt.subplots()
ax.scatter(x_arr,y_arr)
    #  ax.add_artist(lines.Line2D([0,  90], [0, 0],linestyle='--'))
plt.xlabel('',fontsize=18)
    #  plt.suptitle(suptitle,fontsize=16)
plt.title('l-v diagram',fontsize=12)
plt.legend(('',''),loc='best')
plt.show()
```
